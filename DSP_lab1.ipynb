```python
import numpy as np
import matplotlib.pyplot as plt

X_mu = np.array([6, 2, 4, 3, 4, 5, 0, 0, 0, 0])

N = len(X_mu)
k = np.arange(N)
mu = np.arange(N)

K = np.outer(k, mu)
print("Matrix K (Indices Product):")
print(K)

W = np.exp(+1j * 2 * np.pi / N * K)
print("\nMatrix W (Fourier Matrix - first few elements):")
print(np.round(W[:3, :3], 2))

x_k = (1/N) * np.dot(W, X_mu)

plt.figure(figsize=(10, 5))
plt.stem(k, np.real(x_k), linefmt='C0-', markerfmt='C0o', label='Real part')
plt.stem(k, np.imag(x_k), linefmt='C1--', markerfmt='C1x', label='Imaginary part')
plt.title(f'Synthesized Discrete-Time Signal (N={N})')
plt.xlabel('Sample index k')
plt.ylabel('Amplitude x[k]')
plt.legend()
plt.grid(True)
plt.show()

print("\nSynthesized Signal x[k]:")
print(np.round(x_k, 4))
```
